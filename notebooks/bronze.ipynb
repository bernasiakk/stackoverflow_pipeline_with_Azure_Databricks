### Download Files to Azure Blob & unzip them

```python
import requests
import shutil
import py7zr
```

```python
url_prefix = 'https://archive.org/download/stackexchange/'
url_suffix = '.stackexchange.com.7z'

url_list = ['genai', 'health', 'iot', 'lifehacks', 'opendata'] #TODO add whatever categories you'd like

for i in url_list:
    
    url = url_prefix + i + url_suffix
    filename = i + url_suffix
    bronze_dir = '/dbfs/mnt/stackoverflow/bronze'
    file_path = f'{bronze_dir}/{filename}'

    with requests.get(url, stream=True) as r:
        r.raise_for_status()
        with open(file_path, 'wb') as f:
            shutil.copyfileobj(r.raw, f)
    print(f'7zip file downloaded: {file_path}')

    with py7zr.SevenZipFile(file_path, mode='r') as z:
        z.extractall(f'{bronze_dir}/{i}/')

    print(f"Files extracted to {bronze_dir}/{i}/")
```
